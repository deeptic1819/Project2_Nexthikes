#### Fill null values by mean of atemp column :

```python
#Central Tendency value of 'atemp' column.
mean=df['atemp'].mean()
```

```python
# atemp column filled by mean.
df['atemp']=df['atemp'].fillna(mean)
```

```python
df.isnull().sum()
```

#### Drop "instant" column from dataset.

```python
#Remove unnecessary columns (yr) from dataset.
# Syntax: df = df.drop('ColumnName', axis=1)
# Syntax: df.drop(['Column1', 'Columns2'], axis=1)
#df=df.drop(['yr'],axis=1)
```

```python
#Drop unnecessary column from dataset.
df=df.drop(['instant','season','yr'],axis=1)
```

```python
df.head()
```

#### Convert "dteday" to Datetime format.

```python
#df['dteday']=pd.to_datetime(df['dteday'],format= '%d-%m-%Y')
df['dteday']=df['dteday'].astype('datetime64[ns]')
```

#### EDA and Data Visualizations of Dataset :

#### Correlation in data.

```python
# Compute correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
df.head(2)
```

#### Temperature Distribution :

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["temp"], bins=20, kde=True, color="blue")
plt.title("Temperature Configuration")
plt.xlabel("Temperature")
plt.show()
```

#### Hourly Distribution

```python
plt.figure(figsize=(8, 4))
sns.countplot(x=df["hr"], palette="Greens")
plt.title("Hourly Distribution")
plt.xlabel("Hour of the Day")
plt.ylabel("Count")
plt.show()
```

#### Correlation heatmap :

```python
plt.figure(figsize=(10,5))
sns.heatmap(correlation_matrix, annot=True,fmt=".2f",cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

#### Import scipy library to perform Skewness and Kurtosis.

```python
from scipy.stats import kurtosis
```

#### Skewness of the Dataset.

```python
skewness_values = df.select_dtypes(include=['int64', 'float64']).skew()
```

```python
print("Skewness:\n", skewness_values)
```

```python
plt.figure(figsize=(9,4))
sns.barplot(x=skewness_values.index, y=skewness_values.values, palette="dark")
plt.title("Comparison of Skewness in Columns")
plt.xlabel("Columns")
plt.ylabel("Skewness Value")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

#### Calculate Kurtosis.

```python
kurtosis_values = df.select_dtypes(include=['int64', 'float64']).apply(kurtosis)
```

```python
print("\nKurtosis:\n", kurtosis_values)
```

```python
fig, ax = plt.subplots(1,1 , figsize=(8, 4))
sns.barplot(x=kurtosis_values.index, y=kurtosis_values.values, palette="deep")

plt.title("Comparison of Kurtosis in Columns")
plt.xlabel("Columns")
plt.ylabel("Kurtosis Value")

plt.axhline(y=0, color='r', linestyle='--')

plt.tight_layout()
plt.show()
```
